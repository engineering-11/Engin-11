```python
def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx
```

```python
df1 = data_pocket[['deviceTime_unix', 'cpm', 'cpmError']].copy()
df2 = data_d3s[['deviceTime_unix', 'cpm', 'cpmError']].copy()

unix_timerad = df1['deviceTime_unix']
unix_timeweather = df2['deviceTime_unix']

print(unix_timerad.iloc[-1])
print(unix_timeweather.iloc[-1])
```

```python
if(unix_timerad[0]<unix_timeweather[0]):
    start_index = find_nearest(unix_timeweather, unix_timerad[0])
    print("dropping d3s data from 0 - {}".format(start_index))
    df2 = df2.drop(df2.index[0:start_index])
else:
    start_index = find_nearest(unix_timerad, unix_timeweather[0])
    print("dropping pocket data from 0 - {}".format(start_index))
    print("Time range: {} - {}".format(unix_timeweather[0],unix_timerad[start_index]))
    df1 = df1.drop(df1.index[0:start_index])

unix_timerad = df1['deviceTime_unix']
unix_timeweather = df2['deviceTime_unix']

if(unix_timerad.iloc[-1]<unix_timeweather.iloc[-1]):
    stop_index = find_nearest(unix_timerad, unix_timeweather.iloc[-1])
    print("dropping pocket data from {} - {}".format(stop_index,len(unix_timerad)-1))
    df1 = df1.drop(df1.index[stop_index-1:])
else:
    stop_index = find_nearest(unix_timeweather, unix_timerad.iloc[-1])
    print("dropping pocket data from {} - {}".format(stop_index,len(unix_timeweather)-1))
    df2 = df2.drop(df1.index[stop_index:])
    
print(df1['deviceTime_unix'].iloc[0])
print(df2['deviceTime_unix'].iloc[0])
print(df1['deviceTime_unix'].iloc[-1])
print(df2['deviceTime_unix'].iloc[-1])
```
